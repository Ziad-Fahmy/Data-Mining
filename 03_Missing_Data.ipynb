# Missing Data

Let's show a few convenient methods to deal with Missing Data in pandas:

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3],
                  'D': [10, np.nan, np.nan]})
```

```python
df
```

**Check Missing data existance**

```python
df.isna()
```

```python
df.isna().sum(axis=0)
```

```python
df.isna().sum(axis=1)
```

**Droping missing data**

```python
df
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df
```

```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3],
                  'D': [10, np.nan, np.nan]})
```

```python
df
```

```python
df.dropna(axis=1)
```

```python
df
```

```python
df.dropna(thresh=1, axis=0)
```

thresh=1: This parameter specifies the minimum number of non-null values required for a row or column to be retained. <br>
In this case, it is set to 1, meaning that a row will be kept if it has at least one non-null value.

```python
df.dropna(thresh=2)
```

```python
df.dropna(thresh=3)
```

**Fill missing data**

```python
df
```

```python
df.fillna(value=10)
```

```python
df
```

```python
df['A']
```

```python
df['A'].mean()
```

```python
df['A'].fillna(value=df['A'].mean(), inplace=True)
```

```python
df
```

```python
df.fillna(value=df['A'].mean(), inplace=True)
```

```python
df
```

```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3],
                  'D': [10, np.nan, np.nan]})
df
```

```python
df.columns
```

```python
for col in df.columns:
    df[col].fillna(value=df[col].mean(), inplace=True)
df
```
